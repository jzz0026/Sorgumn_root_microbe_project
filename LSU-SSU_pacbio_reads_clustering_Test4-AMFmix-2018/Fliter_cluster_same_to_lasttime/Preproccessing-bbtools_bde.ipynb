```python
#function to return sequence counts of fastq formatted as a data frame
def CountSeqFastq(FilenamePattern, barcode_start, barcode_end, colname): 
    counts = dict()
    barcode_start = int(barcode_start)
    barcode_end = int(barcode_end)
    counter = 0
    for filename in os.listdir(): 
        if filename.endswith(FilenamePattern): 
        #print(filename)
            Barcode = filename[barcode_start:barcode_end]
            handle = open(filename)
            counter = 0
            for line in handle:          
                if line.startswith("+") and line.endswith("+\n"):
                    counter = counter + 1
            NumSeq = counter
            counts[Barcode] = NumSeq
       # print(os.path.join(directory, filename))
        
    df = pd.DataFrame.from_dict(data=counts, orient='index')
    df = df.rename(columns={0: colname})
    
    return df
        
    
```
